```python
import requests
from bs4 import BeautifulSoup
import random
```

```python
l = []
for browser in ['Chrome', 'Firefox', 'Safari', 'Opera']:
    print(browser)
    url = 'http://www.useragentstring.com/pages/useragentstring.php?name={}'.format(browser)
    response = requests.get(url)
    content = response.content
    soup = BeautifulSoup(content)
    
    l += [a.text for a in soup.findAll('a') if a.text[0:7]=='Mozilla']
    print(len(l))
```

```python
random.choice(l)
```
